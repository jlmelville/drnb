[Contrastive Neighbor Embedding](https://github.com/berenslab/contrastive-ne) is a PyTorch-based
framework that can get you very close to a variety of embedding methods, including t-SNE and UMAP.
See [the paper](https://openreview.net/forum?id=B8a1FcY0vi) for more on that.

## Parameters

* `loss_mode`: `"infonce"` (t-SNE), `"neg"` (UMAP) or `"nce"` (NCVis). Default is `"infonce"`.
* `negative_samples`: number of negative samples per positive sample. Default is `5` like UMAP, but
for more t-SNE-like results (hard to describe but more crinkly like the fjords of Norway), you can
set this *much* higher e.g. to `500`.
* `s`: this controls the balance of positive to negative forces. This goes from `0.0` (t-SNE-like)
to `1.0` (UMAP-like). Only applies for `loss_mode="infonce"` or `loss_mode="neg"`.
* `metric`: one of `"euclidean"` or `"cosine"`.
* `batch_size`: by default `32768` (2^15). In terms of memory usage you will need to balance this
with `negative_samples`.
* `n_epochs`: 200 by default.
* `early_exaggeration`: if `True`, the first third of the optimization has `s=1` enforced. Again
only applies if `loss_mode="infonce"` or `loss_mode="neg"`.
* `learning_rate`: default is `0.001`.
* `momentum`: default 0.0. Used for `optimizer="sgd"` only.
* `optimizer`: one of `"sgd"` or `"adam"`. Default is `"sgd"`.
