# PaCMAP: local scaling vs. k-nearest neighbors

PaCMAP uses locally-scaled neighbors by default, but it can also use k-nearest neighbors. Let's see
if there are any datasets where we can see a big difference. Note that we expect to see some
variability in the results just due to the vagaries of the random number generator. See
[pacmap-repeats.ipynb](pacmap-repeats.ipynb) for more details on that.


```python
import drnb.embed.pipeline as pl
from drnb.experiment import Experiment
```

```python
experiment = Experiment(name="pacmap-local-scaling")
```

```python
experiment.add_method(
    "pacmap",
    name="pacmap-lsn",
)

experiment.add_method(
    pl.embedder("pacmap", local_scale=False),
    name="pacmap-knn",
)
```

```python
experiment.add_datasets(
    [
        "curve2d",
        "mammoth",
        "scurvehole",
        "isoswiss",
        "hierarchical",
        "spheres",
        "coil20",
        "coil100",
        "macosko2015",
        "mnist",
        "fashion",
        "ng20pacmap",
    ]
)
```

```python
experiment.run()
```

```python
experiment.plot()
```

```python
experiment.save(overwrite=True)
```
