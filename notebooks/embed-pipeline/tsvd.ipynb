Using truncated SVD *should* give the same results as using `method="pca"`. With default settings, large datasets will use the truncated SVD method under the hood.

```python
import drnb.embed.pipeline as pl

embed_pipeline = pl.create_pipeline(
    method=pl.embedder("tsvd"),
    eval_metrics=["rte", "rpc", "gs", ("nnp", dict(n_neighbors=[15, 50, 150]))],
    verbose=True,
    plot=dict(cex=10, alpha_scale=1, figsize=(6, 4), legend="force"),
)
```

```python
embed_result = embed_pipeline.run("s1k")
```

```python
embed_result["evaluations"]
```

By definition, the "Global Score" assessment (here given as `'gs'`) for PCA is 1. See the [TriMap paper](https://arxiv.org/abs/1910.00204) for more on its definition.
