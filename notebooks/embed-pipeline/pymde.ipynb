PyMDE parameters from the docstring for [preserve_neighbors](https://github.com/cvxgrp/pymde/blob/0e9ab9ddccffb0bdc559cec7f7c614a5c1f0b93d/pymde/recipes.py#L221)

* `attractive_penalty="log1p"`: the attractive part of the cost function. Other penalties from the same module are `"huber"` or `"quadratic"`. 
You should provide a function with the same name as one of the functions in the [pymde.functions.penalties](https://github.com/cvxgrp/pymde/blob/main/pymde/functions/penalties.py)
module.
* `repulsive_penalty="log"`: the repulsive part of the cost function. The pymde documentation says that you should only consider one of
`"log"`, `"invpower"`, or `"logratio"`. If `None` then only positive interactions are considered. 
* `constraint=None`: a constraint that can be used instead of `repulsive_penalty`. Should be one of
`"standardized"` or `"centered"`. If `repulsive_penalty` is not supplied then `"standardized"` is
used. You can learn more about constraints by looking at the [pymde.constraints](https://github.com/cvxgrp/pymde/blob/main/pymde/constraints.py)
module, but in `drnb` only the two options mentioned are supported.
* `n_neighbors=None`: by default, a value between `5-15` is used, depending on an attempt to use ~1% of the total edges in the dataset as positive edges. Practically, this means datasets < 1200 items have `n_neighbors=5`, and datasets > 3000 have `n_neighbors=15`. 
* `repulsive_fraction=None`: fraction of edges (relative to the number of attractive edges) to include in the `repulsive_penalty` calculation. Usually `1`, i.e. the same number of repulsive edges as attractive edges.
* `max_distance=None` If set, the maximum distance a nearest neighbor can have.
* `init="quadratic"` The initialization method. The default, `"quadratic"`, is a Laplacian eigenmap initialization. Otherwise, `"random"` is a random initialization using normal distribution (mean 0, standard deviation 1).

```python
pl.standard_eval(
    method="pymde-plugin",
    dataset="s1k",
    params=dict(
        attractive_penalty="quadratic",
        repulsive_penalty=None,
        constraint="standardized",
    ),
    verbose=True,
)
```
