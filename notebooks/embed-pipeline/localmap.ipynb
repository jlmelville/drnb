[LocalMAP](https://www.arxiv.org/abs/2412.15426) is a variant of PaCMAP that is focused on producing
well-separate clusters. It is available in the [PaCMAP](https://github.com/YingfanWang/PaCMAP)
package from version 0.8 onwards.

```python
import drnb.embed.pipeline as pl

embed_pipeline = pl.create_pipeline(
    method=pl.embedder("localmap"),
    eval_metrics=["rte", "rpc", ("nnp", dict(n_neighbors=[15, 50, 150]))],
    verbose=True,
    plot=dict(cex=10, alpha_scale=1, figsize=(6, 4), legend="force"),
)
```

```python
embed_result = embed_pipeline.run("s1k")
```

If you look at the same results for [PaCMAP](https://github.com/jlmelville/drnb/blob/master/notebooks/embed-pipeline/pacmap.ipynb)
you will see that the clusters really *are* better separated with LocalMAP. But what about the
`mammoth` dataset, beloved of `PaCMAP`, which doesn't really have any clear clusters?

```python
pl.create_pipeline(
    method=pl.embedder(
        "localmap",
    ),
    plot=dict(
        figsize=(9, 6),
        legend=False,
        show_axes=True,
    ),
    eval_metrics=["rte", "rpc", ("nnp", dict(n_neighbors=[15, 50, 150]))],
).run("mammoth")["evaluations"]
```

Not so good here. At any rate the only extra parameter you need to know about is `low_dist_thres`
which determines the distance for what are considered "local" far points. These are points used in
the negative sampling -- by choosing points which are closer than a random sampling, we can get a
better separation of clusters. That said, the value of `low_dist_thres` represents the distance
threshold in the embedded space, so if you decide to change it from the default (`10`), then there
isn't a lot of guidance on what a good value is. Although I have turned off displaying the axes
by default, using the tick labels on the axes to examine the extent of the embedding might help 
here, which is why `show_axes` is set to `True` in the plot configuration.
