```python
import numpy as np
import pandas as pd
```

```python
import drnb.embed.pipeline as pl

embed_pipeline = pl.create_pipeline(
    method=pl.embedder("umap"),
    eval_metrics=["rte", "rpc", "gs", ("nnp", dict(n_neighbors=[15, 50, 150]))],
    verbose=True,
)
```

```python
embed_result = embed_pipeline.run("s1k")
```

UMAP parameters:

* `n_neighbors=15`
* `metric="euclidean"`
* `metric_kwds=None`
* `n_epochs=None`: set to `500` if `N < 10000`, `200` otherwise. If `densmap=True` then an extra 
`200` epochs are added.
* `learning_rate=1.0`
* `init="spectral"`: one of `"spectral"` (spectral embedding like a noisy Laplacian Eigenmap) or 
`"random"` (uniform random sampling between -10 and 10).
* `min_dist=0.1`
* `spread=1.0`
* `n_jobs=-1`
* `set_op_mix_ratio=1.0`
* `local_connectivity=1.0`
* `repulsion_strength=1.0`
* `negative_sample_rate=5`
* `a=None`
* `b=None`
* `random_state=None`: if you set this, then UMAP will not allow for multiple threads to be used,
so expect things to run slower. Usually UMAP will warn you about this, but to avoid spamming the
logs we check and set `n_jobs=1` ahead of time.
* `force_approximation_algorithm=False`: if `True`, then use pynndescent's approximate nearest 
neighbors even for "small" (`N < 4096`) datasets.
* `verbose=False`
* `unique=False`: if `True` then identical points in the input space are represented as one point 
in the output space (which can avoids some trouble with duplicates having zero distance from each 
other).
* `densmap=False`: if `True` then use the `densmap` method.
* `dens_lambda=2.0`: `densmap` parameter.
* `dens_frac=0.3`: `densmap` parameter.
* `dens_var_shift=0.1`: `densmap` parameter.
* `output_dens=False`: `densmap` parameter: if `True`, export the input and output radii.
* `tqdm_kwds=None`: a dict containing configuration for the progress bar when `verbose=True`. 
It doesn't seem to work great in a notebook, so this is disabled by default by `drnb` if 
`verbose=True` is also present (by forcing `tqdm_kwds={"disable": True}`). If you really want tqdm 
output when `verbose=True`, explicitly set `tqdm_kwds` in the `params`.

```python
s1k_densmap = pl.create_pipeline(
    method=pl.embedder(
        "umap",
        dict(densmap=True, output_dens=True),
    ),
    eval_metrics=["rte", "rpc", "gs", ("nnp", dict(n_neighbors=[15, 50, 150]))],
).run("s1k")
```

```python
iris_umap = pl.create_pipeline(
    method=pl.embedder(
        "umap",
        use_precomputed_knn=False,
        params=dict(
            unique=True,
            random_state=42,
            n_neighbors=50,
            verbose=True,
        ),
    ),
    eval_metrics=["rte", "rpc", "gs", ("nnp", dict(n_neighbors=[15, 50, 150]))],
).run("iris")
```
