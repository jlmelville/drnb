NCVis parameters taken from [the Python wrapper](https://github.com/stat-ml/ncvis/blob/master/wrapper/ncvis.pyx) docs.

* `n_threads=-1`: < 1 means use all available CPUs (as calculated by `multiprocessing.cpu_count()`).
* `n_neighbors=15`
* `M=16`: HNSW parameter.
* `ef_construction=200`: another HNSW parameter.
* `random_seed=42`
* `n_epochs=50`
* `n_init_epochs=20`: number of iterations spent during initialization. The NCVis describes this as a power iteration method.
* `spread=1.0`: same as the UMAP parameter.
* `min_dist=0.4`: same as the UMAP parameter.
* `a=None`: same as the UMAP parameter.
* `b=None`: same as the UMAP parameter.
* `alpha=1.0`: learning rate for coordinate updates.
* `alpha_Q=1.0`: learning rate for the normalization factor `Q`.
* `n_noise=None`: The number of noise samples to use per item. UMAP default is `5`. This can also be an array with the number of noise samples to use per per epoch if you want this to vary. `n_epochs` is ignored if this is set, instead the length of the array is used to specify the number of epochs. By default the number of negative samples grows cubically with the epoch number, scaled so that approximately the same total number of noise samples are used as would result if using a fixed value of `n_noise=5`. The result is only approximate because the number of noise samples per item must be at least 1 so rounding up at early iterations means that more than `5 x n_epochs` noise samples can result.
* `distance="euclidean"` one of: (`"euclidean"`, `"cosine"`, `"correlation"`, `"inner_product"`)
