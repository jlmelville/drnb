```python
%load_ext lab_black
%load_ext autotime
import pandas as pd
import numpy as np
```

[Avonet]( https://doi.org/10.111/ele.13898) is a bird morphology dataset with downloadable data at <https://figshare.com/articles/dataset/AVONET_morphological_ecological_and_geographical_data_for_all_birds_Tobias_et_al_2021_Ecology_Letters_/16586228>. There are 11 morphological measurements for over 10,000 bird species. You can see how the 11 traits are defined in Figure 2 of the paper.

```python
import zipfile
from io import BytesIO

import requests

url = "https://figshare.com/ndownloader/files/34480865"
req = requests.get(url, timeout=10)
with zipfile.ZipFile(BytesIO(req.content)) as zip_file:
    with zip_file.open("ELEData/TraitData/AVONET1_BirdLife.csv") as f:
        avonet = pd.read_csv(f)

avonet
```

```python
avonet.columns
```

Here are the 11 physical traits in Avonet:

```python
avonet[
    [
        "Beak.Length_Culmen",
        "Beak.Length_Nares",
        "Beak.Width",
        "Beak.Depth",
        "Tarsus.Length",
        "Wing.Length",
        "Kipps.Distance",
        "Secondary1",
        "Hand-Wing.Index",
        "Tail.Length",
        "Mass",
    ]
]
```

We can do a pair plot to get a rough idea of how correlated these measures are. The mass distribution is quite skewed so I have taken the log:

```python
avonet["logMass"] = np.log(avonet["Mass"])
```

```python
import seaborn as sns

sns.pairplot(
    avonet[
        [
            "Beak.Length_Culmen",
            "Beak.Length_Nares",
            "Beak.Width",
            "Beak.Depth",
            "Tarsus.Length",
            "Wing.Length",
            "Kipps.Distance",
            "Secondary1",
            "Hand-Wing.Index",
            "Tail.Length",
            "logMass",
        ]
    ]
)
```

Seems like we probably don't need the `Nares` *and* the `Culmen` measurements of beak length, but I am no ornithologist, so I'll keep them both. Other traits seem less redundant.

The `Family1` and `Order1` are possible columns you could use for categorizing the birds. `Order1` seems more manageable.

```python
from drnb.dataset import create_data_pipeline

data_pipe = create_data_pipeline(
    data_cols=[
        "Beak.Length_Culmen",
        "Beak.Length_Nares",
        "Beak.Width",
        "Beak.Depth",
        "Tarsus.Length",
        "Wing.Length",
        "Kipps.Distance",
        "Secondary1",
        "Hand-Wing.Index",
        "Tail.Length",
        "logMass",
    ],
    convert=dict(dtype="float32", layout="c"),
    scale="z",
    data_export=["csv", "npy"],
    target_cols=["Species1", "Family1", "Order1"],
    target_export=["csv", "pkl"],
    neighbors=dict(
        n_neighbors=[15, 50, 150],
        method="exact",
        metric=["euclidean"],
        file_types=["csv", "npy"],
    ),
    triplets=dict(
        n_triplets_per_point=5,
        seed=1337,
        file_types=["csv", "npy"],
    ),
    verbose=True,
)
```

```python
data_result = data_pipe.run("avonet", data=avonet, verbose=True)
```
