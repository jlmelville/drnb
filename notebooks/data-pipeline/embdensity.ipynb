```python
%load_ext lab_black
%load_ext autotime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

A 2D dataset consisting of 2 clusters inserted into some noise, from a [google colab notebook by Pavlin Poliƒçar](https://colab.research.google.com/drive/1-8vLka6Cwj8rE4G48w0C7kXqQJLkZrQH?usp=sharing) (originally found in a [tweet](https://twitter.com/pavlinpolicar/status/1633901494784147464)), designed to demonstrate some problems with UMAP and t-SNE distinguishing noise from clusters.

```python
x1 = np.random.uniform(-5, 5, size=(1000, 2))
x2 = np.random.normal((0, 2.5), 0.1, size=(250, 2))
x3 = np.random.normal((0, -2.5), 0.1, size=(250, 2))
x = np.vstack([x1, x2, x3])
y = np.hstack([0 * np.ones(1000), 1 * np.ones(250), 2 * np.ones(250)]).astype(int)

c = np.array(["#666666", "#1f77b4", "#d62728"])[y]

plt.scatter(x[:, 0], x[:, 1], alpha=0.25, s=12, c=c)
```

```python
from drnb.io.pipeline import create_default_pipeline

data_pipe = create_default_pipeline(
    check_for_duplicates=False,
)
```

```python
data_result = data_pipe.run(
    "embdensity",
    data=x,
    target=pd.DataFrame({"cluster": c}),
    url="https://colab.research.google.com/drive/1-8vLka6Cwj8rE4G48w0C7kXqQJLkZrQH",
    tags=["small", "lowdim"],
    verbose=True,
)
```

In a [subsequent tweet](https://twitter.com/pavlinpolicar/status/1633901499049754639) Pavlin notes that DensMAP should be better than UMAP for avoiding giving the appearance of clustering the noise.

```python
import drnb.embed.pipeline as pl
```

```python
pl.standard_eval(
    method="umap",
    dataset="embdensity",
)
```

```python
pl.standard_eval(
    method="umap",
    params=dict(densmap=True),
    dataset="embdensity",
)
```
