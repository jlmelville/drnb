```python
import os
from pathlib import Path

from drnb.embed.context import EmbedContext, read_dataset_from_ctx
from drnb.neighbors.compute import calculate_neighbors

DRNB_HOME = Path(os.environ["DRNB_HOME"]).resolve()
os.environ["DRNB_NN_PLUGIN_KEEP_TMP"] = "1"
```

```python
ctx = EmbedContext(
    dataset_name="s1k",  # matches data/s1k_data.csv.gz
    embed_method_name="annoy-test",
    drnb_home=DRNB_HOME,
)
```

```python
dataset = read_dataset_from_ctx(ctx)
x = dataset[0]
```

```python
nbrs = calculate_neighbors(
    data=x,
    n_neighbors=16,
    metric="euclidean",
    method="annoy",  # this routes through the NN plugin
    verbose=True,
    name=ctx.dataset_name,
    drnb_home=ctx.drnb_home,
    data_sub_dir=ctx.data_sub_dir,
    nn_sub_dir=ctx.nn_sub_dir,
    experiment_name=ctx.experiment_name,
)
```
